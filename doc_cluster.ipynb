```python
from __future__ import print_function
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import os
import numpy as np
```

```python
def create_files(folder_path, filename):
    text_document = open(folder_path + filename, 'r')
    auth_name = filename.replace(".txt","")
#     print auth_name
#     print reviewer_id_dict[auth_name]
    count = 0
    for text_line in text_document: 
        count += 1
        text_line = text_line.rstrip()
        text_line_buffer = text_line.split('|')
        if count < 10:
            fname = str(reviewer_id_dict[auth_name]) +  "0" + str(count)
        else:
            fname = str(reviewer_id_dict[auth_name]) +  str(count)
        new_file = open("./corpus/" + fname, 'w')
        print(text_line_buffer[0],file=new_file, end = "\n")
        print(text_line_buffer[1],file=new_file)
        new_file.close()
```

```python
folder_path = "./title_abstract_data/"
for filename in os.listdir(folder_path):
    create_files(folder_path,filename)
```

```python
folder_path = "./title_abstract_data/"
count = 0
reviewer_id_dict = {}
for filename in os.listdir(folder_path):
    count += 1
    auth_name = filename.replace(".txt","")
#     print (auth_name)
    reviewer_id_dict[auth_name] = count
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
# print(str(os.curdir()))
X = vectorizer.fit_transform(os.listdir("./corpus/"))
# print (X)
# print(vectorizer.get_feature_names())
```

```python
no_clusters = 2
km = KMeans(n_clusters=no_clusters,init='k-means++')
km.fit(X)
clusters = np.ndarray.tolist(km.labels_)
print(len(clusters))
print (clusters)
```

```python
cluster_result_dict = {}
for i in range(no_clusters):
    cluster_result_dict[i] = [k for k,j in enumerate(clusters) if j == i]
```
