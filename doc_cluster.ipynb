```python
def word_tokenizer(text):
            #tokenizes and stems the text
            tokens = word_tokenize(text)
            stemmer = PorterStemmer()
            tokens = [stemmer.stem(t) for t in tokens if t not in stopwords.words('english')]
            return tokens
```

```python
def cluster_sentences_optimal_k():
    print("silhouette score on various k")
    file_object = open("tfidf_res.txt",'r')  
# load the object from the file into var b
    tfidf_matrix = pickle.load(file_object) 

    for i in range (5,25):
        kmeans = KMeans(n_clusters=i)
#         kmeans.fit(tfidf_matrix)
        cluster_labels = kmeans.fit_predict(tfidf_matrix)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
        silhouette_avg = silhouette_score(tfidf_matrix, cluster_labels)
        print("For n_clusters =",i,
          "The average silhouette_score is :", silhouette_avg)
```

```python
def cluster_sentences():
    nb_of_clusters = 24
    file_object = open("tfidf_res.txt",'r')  
    # load the object from the file into var b
    tfidf_matrix = pickle.load(file_object)     
    tfidf_vectorizer = TfidfVectorizer(tokenizer=word_tokenizer,
                                    stop_words=stopwords.words('english'),
                                    lowercase=True,decode_error='ignore')
    #builds a tf-idf matrix for the sentences
    kmeans = KMeans(n_clusters=nb_of_clusters)
    kmeans.fit(tfidf_matrix)
    clusters = collections.defaultdict(list)
    for i, label in enumerate(kmeans.labels_):
            clusters[label].append(i)
    return dict(clusters)
```

```python
def get_best_cluster(filename):
    tfidf_matrix_output = tfidf_matrix[file_no_dict[filename]]
    print(tfidf_matrix_output.shape[1])
    # tfidf_vectorizer = TfidfVectorizer(tokenizer=word_tokenizer,
    #                                     stop_words=stopwords.words('english'),
    #                                     lowercase=True,decode_error='ignore')
    # tfidf_matrix_output = tfidf_vectorizer.fit_transform(content)
#     print(cluster_tfidf_dict[0].shape)
    max = -1
    max_index = 0
    mymat = tfidf_matrix_output.toarray()
    for cluster in range(24):

    #     cosine_similarities = linear_kernel(tfidf_matrix_output[0], cluster_tfidf_dict[cluster]).flatten()
        cosine_similarities = linear_kernel(mymat, cluster_tfidf_dict[cluster]).flatten()
#         print(cosine_similarities)
        if(cosine_similarities > max):
            max = cosine_similarities
            max_index = cluster
    return max_index
#     print(max)
#     print(max_index)
```

```python
def top_expert(cluster_number):
    print(cluster_number)
    cluster_member_count = {}
    for i in cluster_result_dict[cluster_number]:
        if((int(i) %1000) not in cluster_member_count):
            cluster_member_count[(int(i))%1000] = 1
        else:
            cluster_member_count[(int(i))%1000] += 1
    val = cluster_member_count.values()
#     keys = cluster_member_count.keys()
    val_duplicate = cluster_member_count.values()
    val.sort()
    best_expert = []
    for i in range (5):
        print(val_duplicate.index(val[len(val) - i - 1]))
        best_expert.append(val_duplicate.index(val[len(val) - i - 1]))
    return best_expert
```

```python
def get_expert(filename):
    best_cluster = get_best_cluster(filename)
    expert_id = top_expert(best_cluster)
    return getexpert_name(expert_id)
```

```python
def getexpert_name(expert_id):
    expert_list = []
    for id in expert_id:
        for key, value in reviewer_id_dict.iteritems():
    #     print(key, value)
            if value == id + 1:
#                 print (key)
                expert_list.append(key)
                break
    return expert_list
```
